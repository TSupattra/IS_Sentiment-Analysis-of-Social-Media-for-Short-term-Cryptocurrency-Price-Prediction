### Get Path file

```python
# importing required modules
from zipfile import ZipFile
import os
  
def get_all_file_paths(directory):
  
    # initializing empty file paths list
    file_paths = []
  
    # crawling through directory and subdirectories
    for root, directories, files in os.walk(directory):
#         print(files)
        for filename in files:
            # join the two strings in order to form the full filepath.
            filepath = os.path.join(root, filename)
            file_paths.append(filepath)
  
    # returning all file paths
    return file_paths    
```

```python
directory ='./binance_min_price/1month'
Total_filename = get_all_file_paths(directory)
Total_filename_new = []
for i in Total_filename:
    if '.csv' in i:
        Total_filename_new.append(i)
        
Total_filename_new
```

### Concat dataframe

```python
import pandas as pd
directory ='./binance_min_price'
Total_df = pd.DataFrame(columns = ['Open_time','Open', 'High', 'Low', 'Close',  'Volme', 'Close_Time',  'Quote_asset_volumn', 'Number_of_trades', 'buy_base_asset_volumn',  'buy_quote_asset_volumn','empty'])

for i in Total_filename:
    
    df = pd.read_csv(i, header=None)
    Total_df = pd.concat([Total_df,df.rename(columns={0: 'Open_time', 1: 'Open',2: 'High', 3: 'Low',4: 'Close', 5: 'Volme',6: 'Close_Time', 7: 'Quote_asset_volumn', 8: 'Number_of_trades',9: 'buy_base_asset_volumn', 10: 'buy_quote_asset_volumn',11:'empty'})],ignore_index=True)
    
```

### Convert timestamp to datetime

```python
import datetime

def convert_to_datetime(timestamp):
    date = datetime.datetime.fromtimestamp(timestamp / 1e3)
    return date
```

```python
Total_df["Open_time"] = Total_df["Open_time"].apply(convert_to_datetime)


```

```python
Total_df["Close_Time"] = Total_df["Close_Time"].apply(convert_to_datetime)
```

### split Date hour minute second

```python
Total_df['Open_time_dates'] = Total_df['Open_time'].dt.date
Total_df['Open_time_hour'] = Total_df['Open_time'].dt.hour
Total_df['Open_time_minute'] = Total_df['Open_time'].dt.minute
Total_df['Open_time_second'] =Total_df['Open_time'].dt.second
```

```python
Total_df['Close_Time_dates'] = Total_df['Close_Time'].dt.date
Total_df['Close_Time_hour'] = Total_df['Close_Time'].dt.hour
Total_df['Close_Time_minute'] = Total_df['Close_Time'].dt.minute
Total_df['Close_Time_second'] =Total_df['Close_Time'].dt.second
```
