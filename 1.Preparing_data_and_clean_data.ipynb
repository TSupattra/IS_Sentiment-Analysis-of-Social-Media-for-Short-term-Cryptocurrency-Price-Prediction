### Remove Punctuation

```python
df["text_remove_urls_numbers_punctuation"] = df['text_remove_urls_numbers'].str.replace('[^\w\s]','')
df.head()
```

### Removing Stopwords

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
df['text_remove_urls_numbers_punctuation_stopword'] = df['text_remove_urls_numbers_punctuation'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

### Remove duplicate text

```python
df = df.drop_duplicates(subset=['dates','hour','minute','text_remove_urls_numbers_punctuation_stopword'], keep="first")
```
