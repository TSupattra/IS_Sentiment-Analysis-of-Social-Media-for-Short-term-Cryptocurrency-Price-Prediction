```python
def percentage(price_over_time,price_current_time):
    percent = price_over_time-price_current_time
    if percent<0:
        z = -1
    else :
        z = 1
    return z
```

```python
merge_Select_time_close_1hour['Target'] = merge_Select_time_close_1hour.apply(lambda row : percentage(row['Close'],
                     row['Close_original']), axis = 1)
```

### Import vader file

### Find Volume twitter

```python
size = select_vader_df.groupby(['dates','hour']).size().to_frame()
size=size.rename(columns={0: 'Volumn_Twitter'})
```

```python
size
```

### Merge neg neu pos compound and Volumn_Twitter

```python
merge_result_groupBy_min_size = pd.merge(result_groupBy_min,size,on =['dates','hour'],how='outer')
```

```python
merge_result_groupBy_min_size
```

```python
merge_result_groupBy_min_size =merge_result_groupBy_min_size.reset_index()
```

```python
merge_result_groupBy_min_size['hour'] = merge_result_groupBy_min_size['hour'].astype(int)
# merge_result_groupBy_min_size['minute'] = merge_result_groupBy_min_size['minute'].astype(int)
```

```python
merge_result_groupBy_min_size.head()
```

```python
import math
def Geometric_mean(Vpos,Vneg):
    Geometric_mean = math.sqrt(Vpos*Vneg)
    return Geometric_mean
```

```python
merge_result_groupBy_min_size['Geometric_mean_Vpos_Vneg'] = merge_result_groupBy_min_size.apply(lambda row : Geometric_mean(row[('pos', 'mean')],
                     row[('neg', 'mean')]), axis = 1)
```
